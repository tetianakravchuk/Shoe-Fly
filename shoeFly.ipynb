```python

import numpy as np
import pandas as pd

orders = pd.read_csv('orders.csv')

cheap_shoes = orders.groupby('shoe_color').price.apply(lambda x: np.percentile(x, 25)).reset_index()

print(cheap_shoes)
```

```python
# Summary of the data

# Import necessary libraries
import pandas as pd

# Load the dataset
ad_clicks = pd.read_csv('ad_clicks.csv')

# Display the first 10 rows of the dataset
ad_clicks.head(10)

#2. Group by utm_source to Count User Views

# Group by utm_source and count the number of user views
views_by_source = ad_clicks.groupby('utm_source')\
    .user_id.count()\
    .reset_index()

# Display the results
views_by_source

#3. Create is_click Column

# Create a new column `is_click` to check if ad_click_timestamp is not null
ad_clicks['is_click'] = ad_clicks['ad_click_timestamp'].notnull()

# Display the updated DataFrame
ad_clicks.head(10)

#4. Clicks by Source

# Group by utm_source and is_click to count user clicks
clicks_by_source = ad_clicks\
    .groupby(['utm_source', 'is_click'])\
    .user_id.count()\
    .reset_index()

# Display the grouped data
clicks_by_source

#5. Pivot Data for Clicks by Source

# Pivot the data for better readability
clicks_pivot = clicks_by_source\
    .pivot(index='utm_source',
           columns='is_click',
           values='user_id')\
    .reset_index()

# Display the pivoted data
clicks_pivot

#6. Calculate Percentage Clicked

# Calculate the percentage of users who clicked
clicks_pivot['percent_clicked'] = clicks_pivot[True] / \
    (clicks_pivot[True] + clicks_pivot[False])

# Display the updated pivot table
clicks_pivot

#7. Group by Experimental Group

# Group by experimental_group and count the number of users
ad_counts = ad_clicks.groupby('experimental_group').user_id.count().reset_index()

# Display the result
ad_counts

#8. Clicks by Experimental Group

# Group by experimental_group and is_click, then count user clicks
clicks_by_group = ad_clicks.groupby(['experimental_group', 'is_click'])\
    .user_id.count()\
    .reset_index()

# Pivot the data
clicks_pivot = clicks_by_group.pivot(index='experimental_group',
                                     columns='is_click',
                                     values='user_id').reset_index()

# Calculate the percentage of users who clicked
clicks_pivot['percent_clicked'] = clicks_pivot[True] / \
    (clicks_pivot[True] + clicks_pivot[False])

# Display the pivot table
clicks_pivot

#9. Separate Data for Groups A and B

# Filter data for group A
a_clicks = ad_clicks[ad_clicks['experimental_group'] == 'A']

# Filter data for group B
b_clicks = ad_clicks[ad_clicks['experimental_group'] == 'B']

#10. Clicks by Day for Group A

# Group by day and is_click for group A
a_clicks_by_day = a_clicks.groupby(['day', 'is_click'])\
    .user_id.count()\
    .reset_index()

# Pivot the data for better readability
a_clicks_pivot = a_clicks_by_day.pivot(index='day',
                                       columns='is_click',
                                       values='user_id').reset_index()

# Calculate the percentage of users who clicked
a_clicks_pivot['percent_clicked'] = a_clicks_pivot[True] / \
    (a_clicks_pivot[True] + a_clicks_pivot[False])

# Display the results
a_clicks_pivot

#11. Clicks by Day for Group B

# Group by day and is_click for group B
b_clicks_by_day = b_clicks.groupby(['day', 'is_click'])\
    .user_id.count()\
    .reset_index()

# Pivot the data for better readability
b_clicks_pivot = b_clicks_by_day.pivot(index='day',
                                       columns='is_click',
                                       values='user_id').reset_index()

# Calculate the percentage of users who clicked
b_clicks_pivot['percent_clicked'] = b_clicks_pivot[True] / \
    (b_clicks_pivot[True] + b_clicks_pivot[False])

# Display the results
b_clicks_pivot

#Final Outputs

# Display the final results for both groups
print("Group A Clicks by Day:")
print(a_clicks_pivot)

print("\nGroup B Clicks by Day:")
print(b_clicks_pivot)
```
